```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle

# Sample dataset
data = {
    "temperature": [25, 30, 35, 20, 28],
    "humidity": [60, 55, 40, 70, 65],
    "prev_aqi": [80, 120, 150, 60, 100],
    "aqi": [90, 130, 180, 70, 110]
}

df = pd.DataFrame(data)

X = df[["temperature", "humidity", "prev_aqi"]]
y = df["aqi"]

model = LinearRegression()
model.fit(X, y)

# SAVE MODEL
with open("aqi_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("Model saved successfully!")
```
